```python
import random

class RandomizerError(Exception):
    def __init__(self, *args, **kwargs):
        Exception.__init__(self, *args, **kwargs)

def crazy():
    val = random.random()
    if val < 0.1:
        raise ZeroDivisionError('almost divided by 0')
    elif val > 0.9:
        raise ValueError('too large')
    else:
        raise RandomizerError('custom message')
        #return 1 / val

for _ in range(50):
    try:
        print(crazy())
    except ValueError:
        print('value')
    except ZeroDivisionError:
        print('zero')
    except Exception as e:
        print(e)
    finally:
        print('!happens every time!')
```
